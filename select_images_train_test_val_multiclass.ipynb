```python
# Load the CSV file
df = pd.read_csv('Data_Entry_2017_v2020.csv')
df
```

```python
dummy_labels = ['Atelectasis','Effusion', 'Infiltration', 
                'Mass', 'Nodule', 'Pneumothorax', 'No Finding']

# Filter the rows where 'Finding Labels' is in the dummy_labels list.
new_df = df[df['Finding Labels'].isin(dummy_labels)]
```

```python
new_df['Finding Labels'].value_counts()
```

```python
# Randomly select 3,000 records from the 'No Finding' category
no_finding_df = new_df[new_df['Finding Labels'] == 'No Finding']
no_finding_df = no_finding_df.sample(n=3000, random_state=42)

# Randomly select 3,000 records from the 'Infiltration' category
infiltration_df = new_df[new_df['Finding Labels'] == 'Infiltration']
infiltration_df = infiltration_df.sample(n=3000, random_state=42)

# Randomly select 3,00 records from the 'Atelectasis' category
atelectasis_df = new_df[new_df['Finding Labels'] == 'Atelectasis']
atelectasis_df = atelectasis_df.sample(n=3000, random_state=42)

# Randomly select 3,000 records from the 'Effusion' category
effusion_df = new_df[new_df['Finding Labels'] == 'Effusion']
effusion_df = effusion_df.sample(n=3000, random_state=42)
```

```python
# Delete rows labeled 'No Finding' or 'Infiltration' or 'Atelectasis' from new_df
new_df = new_df[~new_df['Finding Labels'].isin(['No Finding', 'Infiltration', 
                                                'Atelectasis', 'Effusion'])]
new_df
```

```python
new_df['Finding Labels'].unique()
```

```python
# Add atelectasis_df, infiltration_df, effusion_df, and nodule_df to the DataFrame new_df
new_df = pd.concat([new_df, no_finding_df, infiltration_df, atelectasis_df, effusion_df])
new_df
```

```python
# Shuffle the data
new_df = new_df.sample(frac=1, random_state=42).reset_index(drop=True)
new_df
```

### Split the data in subsets for train and test

```python
# Import the necessary libraries
import os
import shutil
from sklearn.model_selection import train_test_split

# Define the paths and create folders for training and testing sets
base_path = 'multiclass_x-ray-images/'
all_images_path = 'D:/Lambton/Terms/Term3/AML 3406 - AI and ML Capstone Project/Chest Project/Images/x_ray_images'
```

```python
def create_folders_for_categories(labels, base_path):
    subsets = ['set_pred', 'set_test', 'set_train']

    # Create folders for each subset
    for subset in subsets:
        subset_path = os.path.join(base_path, subset)
        os.makedirs(subset_path, exist_ok=True)

       # Create folders for each disease category in 'set_test', 'set_train' and 'set_pred'
#         if subset != 'set_pred':
        for label in labels:
            category_path = os.path.join(subset_path, label)
            os.makedirs(category_path, exist_ok=True)  # Folder for the category

    print("Folders have been created for each disease category.")
```

```python
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split

def select_images(df, label_category, base_path):
    print(f"Starting the process of selecting images for the {label_category} category...")
    
    df_main = df[df['target'] == label_category]

    # Get subfolder names for train, test, and pred
    train_path, test_path, pred_path = get_sub_folder_names(label_category, base_path)

    # Take a sample of 350 images for set_pred
    df_pred_sample = df_main.sample(350, random_state=42)
    move_images(df_pred_sample, pred_path)

    # Remove the sampled images from df_main
    df_main = df_main.drop(df_pred_sample.index)

    # Split the remaining images into training and testing sets (80% training, 20% testing)
    df_train, df_test = train_test_split(df_main, test_size=0.2, random_state=42)

    # Move disease images to the training and testing sets
    move_images(df_train, train_path)
    move_images(df_test, test_path)

    print(f"Image selection for the {label_category} category has been completed!!")
    print(f"{len(df_train)} images have been selected for the train set.")
    print(f"{len(df_test)} images have been selected for the test set.")
    print(f"{len(df_pred_sample)} images have been selected for set_pred.")
    
#     return df_train, df_test

def move_images(df, target_path):
    for filename in df['img_code']:
        source_path = os.path.join(all_images_path, filename)
        target_file_path = os.path.join(target_path, filename)
        
        os.makedirs(target_path, exist_ok=True)
        shutil.copy(source_path, target_file_path)

def get_sub_folder_names(category_name, base_path):
    train_path = os.path.join(base_path, 'set_train', category_name)
    test_path = os.path.join(base_path, 'set_test', category_name)
    pred_path = os.path.join(base_path, 'set_pred', category_name) #add sub folder for set_pred as well
    
    return train_path, test_path, pred_path
```
