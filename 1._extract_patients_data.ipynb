```python
import pandas as pd
import numpy as np
import cv2
import os

from tqdm import tqdm
```

```python
# Load the CSV file
df = pd.read_csv('Data_Entry_2017_v2020.csv')
df
```

```python
# Get only one random record per patient
# df_patient_group = df.groupby('Patient ID').sample(n=1, random_state=42)
# df_patient_group
```

```python
# Create an empty DataFrame to store the results
df_patient_group = pd.DataFrame()

# Iterate through patient groups
for _, group in df.groupby('Patient ID'):
    # If the group has at least 2 records, select 2 randomly without replacement
    if len(group) >= 2:
        selected_group = group.sample(n=2, random_state=42)
    else:
        # If the group has fewer than 2 records, select only 1 record
        selected_group = group.sample(n=1, random_state=42)
    # Append the selected group to the results DataFrame
    df_patient_group = df_patient_group.append(selected_group)
    
# Reset the indexes of the new DataFrame
df_patient_group = df_patient_group.reset_index(drop=True)
```
